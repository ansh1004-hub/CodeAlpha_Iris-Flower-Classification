# IRIS-FLOWER-CLASSIFICATION USING MACHINE LEARNING.

## INTRODUCTION 
The aim is to classify Iris Flower among three species (setosa,versicolor,verginica) by using measurements of length and width of petals and sepals of flower. 

The iris.data set contains 3 classes of 50 instances each,where each class refers to a type of iris plant.

The central goal here is to design a model that makes useful classifications for new flowers or,one which exhibits good generalization.

The data source is the file iris.data.csv .It contains the data for this example in comma-separated values(CSV) format. The number of column is 5,and the number of rows is 150.

## 
SEPARATING THE INPUT AND THE OUPUT COLUMN 


```python
#Separate features and the target 
data = df.values

X= data[:,0:4]   # Sepal_length,Sepal_width,Petal_length,Petal_width
Y= data[:,4]     # Species 
```

## SPLITTING THE DATA INTO TRAINING AND TESTING 

```python
# Split the data to train and test dataset.
from sklearn.model_selection import train_test_split
X_train , X_test ,Y_train , Y_test = train_test_split(X,Y,test_size = 0.2)


# X_train contains the Input used for training 
# Y_train contains the Output used for training 

# X_test  contains the Input used for test
# Y_test contains the Input used for test
```

## MACHINE LEARNING AND PREDICTIONS 

### MODEL_1 : SUPPORT VECTOR MACHINE ALGORITHMS 

```python
# Support vector machine algorithm
from sklearn.svm import SVC 
model_svc = SVC()
model_svc.fit(X_train,Y_train)
```

```python
# prediction using model1
prediction1 = model_svc.predict(X_test)

# calculating the accuracy of model1
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test,prediction1))
```

### MODEL_2 : LOGISTIC REGRESSION 

```python
# logistic regression 
from sklearn.linear_model import LogisticRegression 
model_LR = LogisticRegression()
model_LR.fit(X_train,Y_train)
```

```python
# prediction using model2
prediction2 = model_LR.predict(X_test)

# calculating the accuracy of model2
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test,prediction2))
```

### MODEL_3 : DECISION TREE CLASSIFIER 

```python
# Decision Tree Classifier 
from sklearn.tree import DecisionTreeClassifier
model_DTC = DecisionTreeClassifier()
model_DTC.fit(X_train,Y_train)
```

```python
# prediction using model3
prediction3 = model_DTC.predict(X_test)

# calculating the accuracy of model3
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test,prediction3))
```

## DEATAILED CLASSIFICATION REPORT TO EVALUATE THE PERFORMANCE OF MACHINE LEARNING CLASSIFICATION MODEL 

```python
# Classification Report 
from sklearn.metrics import classification_report
print(classification_report(Y_test,prediction2))
```

## TESTING THE MACHINE BY GIVING THE NEW DATA 

```python
#Testing 
X_new = np.array([ [3,2,1,0.2] , [4.9,2.2,3.8,1.1] , [5.3,2.5,4.6,1.9] ])   #Input

# Prediction of the species from the input data.
prediction = model_svc.predict(X_new)
print("Prediction of Species: {}".format(prediction))  
```
